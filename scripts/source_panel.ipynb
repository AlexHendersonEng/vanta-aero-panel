```python
import numpy as np
import matplotlib.pyplot as plt
from vanta_aero_panel.panel import Panel, SourcePanelSystem
from vanta_aero_panel.elementary_flows import UniformFlow
from vanta_aero_panel.naca import naca4
```

```python
# Variables
n_panels = 50

# Generate panels
theta = np.linspace(np.pi, -np.pi, n_panels + 1)
x = np.cos(theta)
y = np.sin(theta)

panels = []
for i in range(n_panels):
    panel = Panel(x[i], y[i], x[i + 1], y[i + 1])
    panels.append(panel)

# Panel centres and normals
xc, yc = [panel.xc for panel in panels], [panel.yc for panel in panels]
xn, yn = [panel.xn for panel in panels], [panel.yn for panel in panels]

# Plot panels
padding = 0.25
plt.figure()
plt.plot(
    x, y,
    linewidth=2, linestyle='-', color='blue', label='Panels'
)
plt.plot(
    xc, yc,
    linestyle='none', color='blue',
    marker='o', markersize=5, markerfacecolor='black', markeredgecolor='black', label='Panel Centres'
)
plt.quiver(xc, yc, xn, yn, angles='xy', scale_units='xy', scale=5, label='Normals')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([np.min(x) - padding, np.max(x) + padding])
plt.ylim([np.min(y) - padding, np.max(y) + padding])
plt.grid()
plt.legend()
plt.show()
```

```python
# Variables
u_inf = 1.0
alpha = 0.0

# Generate grid
N = 100
x_start, x_end = -4.0, 4.0
y_start, y_end = -2.0, 2.0
x_array = np.linspace(x_start, x_end, N)
y_array = np.linspace(y_start, y_end, N)
x_grid, y_grid = np.meshgrid(x_array, y_array)

# Create uniform flow
freestream = UniformFlow(u_inf, alpha)

# Create source panel system
source_panels = SourcePanelSystem(panels, freestream)

# Solve
source_panels.solve()

# Velocity field
u_grid, v_grid = source_panels.velocity(x_grid, y_grid)

# Plot streamlines
plt.figure()
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.streamplot(x_grid, y_grid, u_grid, v_grid,
               density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.plot(
    x, y,
    linewidth=2, linestyle='-', color='black', label='Panels'
)
plt.plot(
    xc, yc,
    linestyle='none', color='blue',
    marker='o', markersize=5, markerfacecolor='black', markeredgecolor='black', label='Panel Centres'
)
plt.grid()
plt.axis('scaled')
plt.legend()
plt.show()

# Compute pressure coefficient field
cp = 1.0 - (u_grid ** 2 + v_grid ** 2) / freestream.u_inf ** 2

# Plot pressure coefficient field
plt.figure()
plt.xlabel('x (m)')
plt.ylabel('y (m)')
contf = plt.contourf(x_grid, y_grid, cp,
                     levels=np.linspace(-2.0, 1.0, 100), extend='both')
cbar = plt.colorbar(contf)
cbar.set_label('$C_p$')
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
plt.plot(
    x, y,
    linewidth=2, linestyle='-', color='black', label='Panels'
)
plt.plot(
    xc, yc,
    linestyle='none', color='blue',
    marker='o', markersize=5, markerfacecolor='black', markeredgecolor='black', label='Panel Centres'
)
plt.grid()
plt.axis('scaled')
plt.legend()
plt.show()

# Get each panels theta angle
thetas = np.atan2(yc, xc)
thetas = np.where(thetas < 0, thetas + 2 * np.pi, thetas)

# Sort by theta
idx = np.argsort(thetas)
thetas_sorted = thetas[idx]
cp_sorted = np.array(source_panels.cp)[idx]

# Plot coefficient of pressure over cylinder surface
plt.figure()
plt.xlabel('theta (degrees)')
plt.ylabel('$C_p$')
plt.plot(np.degrees(thetas_sorted), cp_sorted)
plt.grid()
plt.show()
```
