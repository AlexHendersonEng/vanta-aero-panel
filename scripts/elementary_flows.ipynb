```
import numpy as np
import matplotlib.pyplot as plt
from vanta_aero_panel.elementary_flows import Source, UniformFlow, Doublet, Vortex
from scipy import optimize
```

```
# Generate grid
N = 200
x_start, x_end = -4.0, 4.0
y_start, y_end = -2.0, 2.0
x_array = np.linspace(x_start, x_end, N)
y_array = np.linspace(y_start, y_end, N)
x, y = np.meshgrid(x_array, y_array)
```

```
# Create freestream
u_inf = 1.0
alpha = 0.0
freestream = UniformFlow(u_inf, alpha)

# Create source
x_source, y_source = 0.0, 0.0
strength_source = 5.0
source = Source(strength_source, x_source, y_source)

# Superposition
def velocity(xy):
    u1, v1 = freestream.velocity(xy[0], xy[1])
    u2, v2 = source.velocity(xy[0], xy[1])

    return [
        u1 + u2,
        v1 + v2
    ]

def stream_function(xy):
    psi1 = freestream.stream_function(xy[0], xy[1])
    psi2 = source.stream_function(xy[0], xy[1])

    return psi1 + psi2

u, v = velocity([x, y])
psi = stream_function([x, y])

# Stagnation point
x_stagnation = x_source - source.strength / (2 * np.pi * freestream.u_inf)
y_stagnation = y_source

# Contours
levels = [freestream.stream_function(x_stagnation, y_stagnation) - source.stream_function(x_stagnation, y_stagnation),
          freestream.stream_function(x_stagnation, y_stagnation) + source.stream_function(x_stagnation, y_stagnation)]

# Plot streamlines
plt.figure()
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.streamplot(x, y, u, v,
               density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_source, y_source,
            color='#FF0000', s=80, marker='o', label='source')
plt.scatter(x_stagnation, y_stagnation,
            color='#00FF00', s=80, marker='o', label='stagnation')
plt.contour(x, y, psi,
            levels=levels,
            linewidths=2, linestyles='solid', colors='#000000')
plt.grid()
plt.legend()
plt.show()

# Compute pressure coefficient field
cp = 1.0 - (u ** 2 + v ** 2) / u_inf ** 2

# Plot pressure coefficient field
plt.figure()
plt.xlabel('x (m)')
plt.ylabel('y (m)')
contf = plt.contourf(x, y, cp,
                     levels=np.linspace(-2.0, 1.0, 100), extend='both')
cbar = plt.colorbar(contf)
cbar.set_label('$C_p$')
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
plt.scatter(x_source, y_source,
            color='#FF0000', s=80, marker='o', label='source')
plt.scatter(x_stagnation, y_stagnation,
            color='#00FF00', s=80, marker='o', label='stagnation')
plt.contour(x, y, psi,
            levels=levels,
            linewidths=2, linestyles='solid', colors='#000000')
plt.grid()
plt.legend()
plt.show()
```

```
# Create freestream
u_inf = 1.0
alpha = 0.0
freestream = UniformFlow(u_inf, alpha)

# Create source
x_source, y_source = -1.0, 0.0
strength_source = 5.0
source = Source(strength_source, x_source, y_source)

# Create sink
x_sink, y_sink = 1.0, 0.0
strength_sink = -5.0
sink = Source(strength_sink, x_sink, y_sink)

# Superposition
def velocity(xy):
    u1, v1 = freestream.velocity(xy[0], xy[1])
    u2, v2 = source.velocity(xy[0], xy[1])
    u3, v3 = sink.velocity(xy[0], xy[1])

    return [
        u1 + u2 + u3,
        v1 + v2 + v3
    ]

def stream_function(xy):
    psi1 = freestream.stream_function(xy[0], xy[1])
    psi2 = source.stream_function(xy[0], xy[1])
    psi3 = sink.stream_function(xy[0], xy[1])

    return psi1 + psi2 + psi3

u, v = velocity([x, y])
psi = stream_function([x, y])

# Stagnation points
sol1 = optimize.root(velocity, [2.0 * source.x, 2.0 * source.y])
sol2 = optimize.root(velocity, [2.0 * sink.x, 2.0 * sink.y])

x_stagnation = [sol1.x[0], sol2.x[0]]
y_stagnation = [sol1.x[1], sol2.x[1]]

# Contours
levels = [stream_function([x_stagnation[0], y_stagnation[0]])]

# Plot streamlines
plt.figure()
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.streamplot(x, y, u, v,
               linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_source, y_source,
            color='#FF0000', s=80, marker='o', label='source')
plt.scatter(x_sink, y_sink,
            color='#FF0000', s=80, marker='o', label='sink')
plt.scatter(x_stagnation, y_stagnation,
            color='#00FF00', s=80, marker='o', label='stagnation')
plt.contour(x, y, psi,
            levels=levels,
            linewidths=2, linestyles='solid', colors='#000000')
plt.grid()
plt.legend()
plt.show()

# Compute pressure coefficient field
cp = 1.0 - (u ** 2 + v ** 2) / u_inf ** 2

# Plot pressure coefficient field
plt.figure()
plt.xlabel('x (m)')
plt.ylabel('y (m)')
contf = plt.contourf(x, y, cp,
                     levels=np.linspace(-2.0, 1.0, 100), extend='both')
cbar = plt.colorbar(contf)
cbar.set_label('$C_p$')
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
plt.scatter(x_source, y_source,
            color='#FF0000', s=80, marker='o', label='source')
plt.scatter(x_sink, y_sink,
            color='#FF0000', s=80, marker='o', label='sink')
plt.scatter(x_stagnation, y_stagnation,
            color='#00FF00', s=80, marker='o', label='stagnation')
plt.contour(x, y, psi,
            levels=levels,
            linewidths=2, linestyles='solid', colors='#000000')
plt.grid()
plt.legend()
plt.show()
```

```
# Create freestream
u_inf = 1.0
alpha = 0.0
freestream = UniformFlow(u_inf, alpha)

# Create doublet
x_doublet, y_doublet = 0.0, 0.0
strength_doublet = 5.0
doublet = Doublet(strength_doublet, x_doublet, y_doublet)

# Compute radius
radius = np.sqrt(doublet.strength / (freestream.u_inf * 2 * np.pi))

# Superposition
def velocity(xy):
    u1, v1 = freestream.velocity(xy[0], xy[1])
    u2, v2 = doublet.velocity(xy[0], xy[1])

    return [
        u1 + u2 ,
        v1 + v2
    ]

def stream_function(xy):
    psi1 = freestream.stream_function(xy[0], xy[1])
    psi2 = doublet.stream_function(xy[0], xy[1])

    return psi1 + psi2

u, v = velocity([x, y])
psi = stream_function([x, y])

# Stagnation points
x_stagnation = radius * np.array([1.0, -1.0])
y_stagnation = [0.0, 0.0]

# Contours
levels = [stream_function([x_stagnation[0], y_stagnation[0]])]

# Plot streamlines
plt.figure()
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.streamplot(x, y, u, v,
               linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_doublet, y_doublet,
            color='#FF0000', s=80, marker='o', label='doublet')
plt.scatter(x_stagnation, y_stagnation,
            color='#00FF00', s=80, marker='o', label='stagnation')
plt.contour(x, y, psi,
            levels=levels,
            linewidths=2, linestyles='solid', colors='#000000')
plt.grid()
plt.legend()
plt.show()

# Compute pressure coefficient field
cp = 1.0 - (u ** 2 + v ** 2) / u_inf ** 2

# Plot pressure coefficient field
plt.figure()
plt.xlabel('x (m)')
plt.ylabel('y (m)')
contf = plt.contourf(x, y, cp,
                     levels=np.linspace(-2.0, 1.0, 100), extend='both')
cbar = plt.colorbar(contf)
cbar.set_label('$C_p$')
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
plt.scatter(x_doublet, y_doublet,
            color='#FF0000', s=80, marker='o', label='doublet')
plt.scatter(x_stagnation, y_stagnation,
            color='#00FF00', s=80, marker='o', label='stagnation')
plt.contour(x, y, psi,
            levels=levels,
            linewidths=2, linestyles='solid', colors='#000000')
plt.contour(x, y, psi,
            levels=levels,
            linewidths=2, linestyles='solid', colors='#000000')
plt.grid()
plt.legend()
plt.show()

# Compute pressure coefficient over the surface of the cylinder
theta = np.linspace(0, 2 * np.pi, 100)
x_surface = radius * np.cos(theta)
y_surface = radius * np.sin(theta)
u_surface, v_surface = velocity([x_surface, y_surface])
cp_surface = 1.0 - (u_surface ** 2 + v_surface ** 2) / u_inf ** 2

# Plot pressure coefficient over the surface of cylinder
plt.figure()
plt.xlabel('theta (degrees)')
plt.ylabel('$C_p$')
plt.plot(np.degrees(theta), cp_surface)
plt.grid()
plt.show()

# Calculate coefficient of lift and drag
cl = -np.trapezoid(cp_surface * np.sin(theta), theta)
cd = -np.trapezoid(cp_surface * np.cos(theta), theta)
print(f"Coefficient of lift: {cl}")
print(f"Coefficient of drag: {cd}")
```
